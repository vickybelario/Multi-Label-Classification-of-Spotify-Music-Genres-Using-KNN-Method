```python
import ast

genres = []
for g in df['artist_genres']:
    # Safely evaluate the string as a list
    genre_list = ast.literal_eval(g)
    genres.extend(genre_list)  # Add the genres to the list

# Remove duplicates by converting to a set and back to a list
unique_genres = list(set(genres))

unique_genres
```

```python
import ast

# Initialize a list of lists for each genre
gen_mv = [[] for i in range(len(unique_genres))]

for dat in df['artist_genres']:
    # Safely evaluate the string as a list
    genres = ast.literal_eval(dat)
    
    for i, g in enumerate(unique_genres):
        if g in genres:
            gen_mv[i].append(1)
        else:
            gen_mv[i].append(0)

# Convert the result into a DataFrame
gen_mv_dat = pd.DataFrame(np.array(gen_mv).T, columns=unique_genres)

# Display the DataFrame
print(gen_mv_dat)
```

```python
title_df = df[['track_name']].reset_index(drop=True)
song_vector = pd.concat([title_df,gen_mv_dat],axis=1)
song_vector.set_index('track_name',inplace=True)
song_vector
```

```python
def cosine_sim(vect1,vect2):
  norm_1 = np.linalg.norm(vect1)
  norm_2 = np.linalg.norm(vect2)

  cos_sim = (vect1 @ vect2) / (norm_1 * norm_2)
  return cos_sim
```

```python
def recsys(song, top_N):
  cossim = pd.Series([cosine_sim(song_vector.loc[song],x) for x in song_vector.values],index=song_vector.index).drop(index=song)
  print(f'You like {song}, so based on our recommender system, We recommend you to listen:')
  for i,mv in enumerate(cossim.sort_values(ascending=False)[:top_N].index):
    print(f'{i+1}. {mv}')
```
